# Importing necessary libraries



# Note

> This is a synthetic dataset used to depict what the final output generated by the 3 models would look like. It is meant to serve as a foundational component in the development of the minimum viable product.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading the dataset

```python
df=pd.read_csv('/content/syntheticdata.csv')
df.head()
```

# Exploratory Data Analysis

```python
# first few rows of the dataset
df.head(10)
```

```python
# statistics
df.describe()
```

```python
# rows in the dataset
df.info()
```

```python
# checking for null values
print(df.isnull().sum(), '\n')
print('\n','***'*20)
print("total null values in the dataset: ", df.isnull().sum().sum())

# checking for duplicated values
print('\n','***'*20)
print("\ntotal duplicated values in the dataset: ", df.duplicated().sum())
```

```python
# dropping null values
df=df.dropna()
```

```python
# unique values in categorical columns
for column in df.select_dtypes(include=['object']):
    print(f"{column}: {df[column].nunique()} \n")
```

## Visualization

```python
# numerical columns
num=[column for column in df.columns if df.dtypes[column]!='object']
num.pop(0); #excluding student id column
# excluding the test fee column as it is a constant
num.pop(4)
```

```python
# plotting histograms for numerical columns to check distributions
df[num].hist(bins=20, figsize=(10, 6))
plt.suptitle('Histogram of numerical columns');
plt.tight_layout()
```

```python
# plotting boxplot to identify outliers
df[num].plot(kind='box', vert=False, figsize=(10, 6))
plt.title('Box plot of Numerical Columns')
plt.show()
```

```python
# plotting pairplot to determine relationships between different numerical columns
plt.figure(figsize=(6, 6))
pairplot = sns.pairplot(df[num], height= 2, aspect=1)
plt.suptitle('Pairplot of Numerical Columns', y=1.02)
plt.show()
```

```python
# calculating the correlation matrix
correlation_matrix = df[num].corr()

# plotting the heatmap to understand correlation between columns
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='twilight', fmt=".2f")
plt.title('Heatmap of Numerical Columns')
plt.show()
```
