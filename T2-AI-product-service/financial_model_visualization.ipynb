```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# defining base revenue and initial quantities
x=100 # assuming 100 people subscribed the basic plan
y=80 # assuming 80 people subscribed the standard plan
z=50 # assuming 50 people subscribed the premium plan
n=4 #considering a 4 member family subscription
f=10 #total families
fam_var= (200 + n*200)*f #200 is the set base price
base_revenue = 300*x + 500*y + 800*z + fam_var # Assuming some initial quantities

# defining growth rate
growth_rate = 0.25  # rough estimate, assuming the market grows 25% per year

# defining time range
time_range = np.linspace(0, 10, 100)  # simulates 10 years with 100 data points

# defining penetration rate (positive for increase, negative for decrease)
penetration_rate = 0.05  # 40% penetration rate of the app in the market

def total_revenue(t):
  """
  Calculates total revenue at a given time point 't'.
  """
  return base_revenue * (1 + growth_rate) ** t * (1 + penetration_rate)

# calculating total revenue for each time step
total_revenue_values = total_revenue(time_range)

# plotting the total revenue vs time
plt.plot(time_range, total_revenue_values, label='Total Revenue')
plt.xlabel('Time (Years)')
plt.ylabel('Total Revenue')
plt.title('Total Revenue over Time')
plt.legend()
plt.show()
```
