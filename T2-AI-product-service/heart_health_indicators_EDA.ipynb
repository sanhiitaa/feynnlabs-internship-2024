Ploted histogram for each attribute of the dataset.

```python
# selecting non-binary columns
columns=['Income', 'Age',  'Education','PhysHlth', 'MentHlth', 'GenHlth', 'BMI' ]

#setting subplots
fig,axes=plt.subplots(7,1,figsize=(15,20))

# initializing loop to plot over the selected columns
for i, cols in enumerate(columns):
  ax=axes.flat[i]
  df[cols].value_counts().plot(kind='bar',ax=ax)
  ax.set_xlabel(f'column: {cols}')
plt.suptitle('Non-Binary Columns Distribution', y=1)
plt.tight_layout()
plt.show()
```

All the features except plotted above('Income', 'Age',  'Education','PhysHlth', 'MentHlth', 'GenHlth', 'BMI') are binary columns.
1. For Income attribute most people in the Dataset belong to 8th category i.e Age 55 to 59.
2. For Age attribute most people in the Dataset belong to 9th category Age 60 to 64.
3. For Education attribute most people in the Dataset belong to 6th category i.e Post graduate or higher degree. 
4. The most common BMI of the people of this dataset is 27.
5. Most number of people in the dataset have not been suffering from bad physical health from last 30 days.
6. Most number of people in the dataset have not been suffering from bad mental health from last 30 days.
7. Most number of people in the dataset falls under very good general health category.

```python
# # Plot the correlation matrix

plt.figure(figsize=(15, 10))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, linewidth=2, cmap='viridis', fmt=".1f")
plt.show()
```

Plotting the heatmap for all attributes.
the colors indicate the strength and direction of correlation between various attributes.
Darker purple indicates a stronger negative correlation, while brighter yellow suggests a stronger positive correlation.
There’s a strong positive correlation between “HighBP” and “HeartDiseaseorAttack”, and a negative correlation between “PhysicalActivity” and “Age”.
