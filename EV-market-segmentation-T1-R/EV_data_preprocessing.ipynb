```python
# dropping the row with grand total values
final_dataset.drop(32, inplace=True)

# dropping columns with more than 25% null values and the repeated state column
drop_cols=['state']
final_dataset=final_dataset.drop(columns=drop_cols)
```

After analyzing the `Grand total` column, it appears that the NaN values in the other columns can be safely replaced with 0. This conclusion is drawn based on the observation that the `Grand total` remains consistent despite the presence of missing values in other columns.


```python
cols_list=['Special Category Vehicles','Ambulance/Hearses','Construction Equipment Vehicle','Other', 'Three Wheeler','Goods Vehicles','Public Service Vehicle']
final_dataset[cols_list]=final_dataset[cols_list].fillna(0)
```
